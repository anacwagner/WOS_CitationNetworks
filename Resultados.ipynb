### 2.4 - Listando Todas as Citações Realizadas por Cada Artigo

```python
print('2.4 - Listando Todas as Citações Realizadas por Cada Artigo')

# ALTERAR O INTERVALO ANTES DE RODAR
for journal_id in range(6,10):
    
    # REINICIA OS DATAFRAMES
    journals_cit = pd.read_csv('outputs/journals_cit.csv', index_col=0)
    articles_cit = pd.read_csv('outputs/articles_cit.csv', index_col=0)
    citations = pd.DataFrame(columns = ['SOURCE_A', 'SOURCE_J', 'TARGET_A', 'TARGET_J', 'YEAR'])
    
    aux = articles2[articles2.JOURNAL_ID == journal_id]
    n = len(aux)
    
    ID_A_START = aux.index[0]
    i = ID_A_START
    
    while ID_A_START < i+n:
        
        print('Início da Listagem de Todas as Citações Realizadas pelo Artigo ' + str(ID_A_START) + ' Publicado pelo Periódico ' + str(journal_id))
        
        driver = webdriver.Chrome(executable_path = r'/Users/Walter/Documents/chromedriver')
        APPS = 'http://apps-webofknowledge.ez91.periodicos.capes.gov.br'
        driver.get( APPS + '/WOS_GeneralSearch.do')
        
        # REFAZ A BUSCA AVAÇADA PARA RESTABELECER OS LINKS
        advancedSearch(journals, year_list)
        
        articles_cit, journals_cit, citations, ID_A_STOP = allCitations(aux, journals2, ID_A_START, articles_cit, journals_cit, citations)

        # A CADA ETAPA SALVA OS RESULTADOS EM CSV
        articles_cit.to_csv('outputs/articles_cit_ID_J_%s.csv' % journal_id)
        journals_cit.to_csv('outputs/journals_cit_ID_J_%s.csv' % journal_id)
        citations.to_csv('outputs/citations_ID_J_%s.csv' % journal_id)

        print('Citações Realizadas pelos Artigos do Periódico ' + str(journal_id) + ' Armazenadas!')   
        
        ID_A_START = ID_A_STOP + 1
        
        driver.close()
        driver.quit()
        
    print('Citações Realizadas pelos Artigos do Periódico ' + str(journal_id) + ' Extraídas!')
```

### 2.5 - Rede de Citações Finalizada

Na etapa anterior, para cada journal foram gerados três arquivos `csv`. Por exemplo, para o **journal 1**, temos: `articles_cit_ID_J_1`, `journals_cit_ID_J_1` e `citations_ID_J_1`. 

Agora, vamos juntar todos os arquivos com as citações em uma mesma tabela. Para as tabelas `articles_cit` e `journals_cit` vamos ter que retirar as duplicatas e somar os resultados das demais colunas. 

#### `citations`

```python
citations = pd.read_csv('outputs/citations_ID_J_1.csv', index_col=0)

for journal_id in range(2,60):

    aux = pd.read_csv('outputs/citations_ID_J_%s.csv' % journal_id, index_col=0)
    citations = citations.append(aux, ignore_index=True)

citations.index = citations.index + 1
citations.to_csv('outputs/citations.csv')

# citations = pd.read_csv('outputs/citations.csv', index_col=0)
citations
```

#### `articles_cit`

```python
articles_cit = pd.read_csv('outputs/articles_cit_ID_J_1.csv', index_col=0)

for journal_id in range(2,60):

    articles_set = set(articles_cit.index)
    aux = pd.read_csv('outputs/articles_cit_ID_J_%s.csv' % journal_id, index_col=0)
    aux_set = set(aux.index) 
    
    new_article = aux_set - articles_set
    rep_article = articles_set - (articles_set - aux_set) 
    
    # Inserir os novos índices (WOS)
    for i in new_article:
        articles_cit.loc[i] = [aux.TITLE[i], aux.JOURNAL_ID[i], aux.YEAR[i], aux.CITED_IN[i], aux.CITED_OUT[i]]
        
    # Atualizar as colunas dos índices (WOS) repetidos
    for j in rep_article:
        articles_cit.CITED_IN[j] = articles_cit.CITED_IN[j] + aux.CITED_IN[j]
        articles_cit.CITED_OUT[j] = articles_cit.CITED_OUT[j] + aux.CITED_OUT[j]
    
    articles_cit.to_csv('outputs/articles_cit_2.csv')
    print('Journal ' + str(journal_id))

articles_cit.to_csv('outputs/articles_cit_2.csv')

# articles_cit = pd.read_csv('outputs/articles_cit.csv', index_col=0)
articles_cit
```
